```python
class NeuralNetwork:
    
    def __init__(self):
        
        self.wall = game.W.w

        
    def DistanceAndColorInputs(self):
        
        self.starve1 = game.s1.starve
        self.starve2 = game.s2.starve
        
        self.vx1 = game.s1.vx
        self.vy1 = game.s1.vy
        self.vx2 = game.s2.vx
        self.vy2 = game.s2.vy
        
        self.snake1 = game.s1.body
        self.snake2 = game.s2.body
        self.head1 = self.snake1[0].coords
        self.head2 = self.snake2[0].coords
        
        self.numy = game.F1.numy, game.F2.numy
        self.yuck = game.F1.yuck, game.F2.yuck
        
        self.distances1, self.colors1 = self.LookAround(self.head1)
        #self.distances2, self.colors2 = self.LookAround(self.head2)
        
    def LookAround(self, head):
        
        found = [False] * 8
        x = [head[0]] * 8
        y = [head[1]] * 8
        distance = [0] * 8
        color = [0] * 8
        
        while True:
            
            if found[0] == False:
                x[0] += 1
                X, Y, C, found[0] = self.CheckingSquareForLifeSigns(x[0], y[0])
                
                if found[0] == True:
                    distance[0] = X
                    color[0] = C
                    
            if found[1] == False:
                x[1] -= 1
                X, Y, C, found[1] = self.CheckingSquareForLifeSigns(x[1], y[1])
                
                if found[1] == True:
                    distance[1] = X
                    color[1] = C
                    
            if found[2] == False:
                y[2] += 1
                X, Y, C, found[2] = self.CheckingSquareForLifeSigns(x[2], y[2])
                
                if found[2] == True:
                    distance[2] = Y
                    color[2] = C
            
            if found[3] == False:
                y[3] -= 1
                X, Y, C, found[3] = self.CheckingSquareForLifeSigns(x[3], y[3])
                
                if found[3] == True:
                    distance[3] = Y
                    color[3] = C
            
            if found[4] == False:
                x[4] += 1
                y[4] += 1
                X, Y, C, found[4] = self.CheckingSquareForLifeSigns(x[4], y[4])
                
                if found[4] == True:
                    distance[4] = X + Y
                    color[4] = C
                
            if found[5] == False:
                x[5] -= 1
                y[5] += 1
                X, Y, C, found[5] = self.CheckingSquareForLifeSigns(x[5], y[5])
                
                if found[5] == True:
                    distance[5] = X + Y
                    color[5] = C
                
            if found[6] == False:
                x[6] += 1
                y[6] -= 1
                X, Y, C, found[6] = self.CheckingSquareForLifeSigns(x[6], y[6])
                
                if found[6] == True:
                    distance[6] = X + Y
                    color[6] = C
                
            if found[7] == False:
                x[7] -= 1
                y[7] -= 1
                X, Y, C, found[7] = self.CheckingSquareForLifeSigns(x[7], y[7])
                
                if found[7] == True:
                    distance[7] = X + Y
                    color[7] = C
            
            if found[0] == found[1] == found[2] == found[3] == found[4] == found[5] == found[6] == found[7] == True:
                break
                
        return distance, color
        
    def CheckingSquareForLifeSigns(self, x, y):

        distanceX = 0
        distanceY = 0
        color = 0
        found = False
        
        if (x, y) in list(map(lambda z:z.coords, self.wall)):
            distanceX = abs(x - self.head1[0])
            distanceY = abs(y - self.head1[1])
            color = sum(colors.black) // 3
            found = True

        elif (x, y) in list(map(lambda z:z.coords, self.snake2[0:])):
            distanceX = abs(x - self.head1[0])
            distanceY = abs(y - self.head1[1])
            color = sum(colors.silver) // 3
            found = True

        elif (x, y) in list(map(lambda z:z.coords, self.snake1[0:])):
            distanceX = abs(x - self.head1[0])
            distanceY = abs(y - self.head1[1])
            color = sum(colors.gold) // 3
            found = True

        elif (x, y) in list(self.numy):
            distanceX = abs(x - self.head1[0])
            distanceY = abs(y - self.head1[1])
            color = sum(colors.red) // 3
            found = True

        elif (x, y) in list(self.yuck):
            distanceX = abs(x - self.head1[0])
            distanceY = abs(y - self.head1[1])
            color = sum(colors.cyan) // 3
            found = True

        return distanceX, distanceY, color, found
                                
```
